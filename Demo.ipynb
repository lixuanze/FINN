```python
import numpy as np
import tensorflow as tf

activation = tf.tanh
ann = tf.keras.Sequential(
    layers=[tf.keras.layers.Dense(50, activation=activation, input_shape=(2,))] + \
           [tf.keras.layers.Dense(50, activation=activation) for i in range(1, 2)] + \
           [tf.keras.layers.Dense(1, activation=tf.keras.activations.softplus)],
    name="ann"
)
ann.load_weights('gbm_call_trained_models/gbm_european_call_0.125_496.h5')

input_data = np.array([[1.1, 0.36]])

inputs_symbolic = tf.keras.Input(shape=(2,))

x = inputs_symbolic
intermediate_outputs = []
for layer in ann.layers:
    x = layer(x)
    intermediate_outputs.append(x)

activation_model = tf.keras.models.Model(inputs=inputs_symbolic, outputs=intermediate_outputs)

activations = activation_model.predict(input_data)

for i, activation in enumerate(activations):
    print(f"Output of layer {i + 1}: {activation}")
```

```python
import numpy as np
import tensorflow as tf

activation = tf.tanh
ann = tf.keras.Sequential(
    layers=[tf.keras.layers.Dense(50, activation=activation, input_shape=(2,))] + \
           [tf.keras.layers.Dense(50, activation=activation) for i in range(1, 2)] + \
           [tf.keras.layers.Dense(1, activation=tf.keras.activations.softplus)],
    name="ann"
)
ann.load_weights('gbm_call_trained_models/gbm_european_call_0.125_496.h5')

S = tf.Variable([[1.1, 0.36]], dtype=tf.float32)

with tf.GradientTape() as tape:
    tape.watch(S)

    output = ann(S)

delta = tape.gradient(output, S)[:, 0]

print("Delta:", delta.numpy())
```

```python
import scipy.stats as si
def bs(S, K, T, r, sigma):
        d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
        d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
        call = (S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0))
        call_delta = si.norm.cdf(d1)
        call_theta = - (S * sigma * si.norm.pdf(d1)) / (2 * np.sqrt(T)) - r * K * np.exp(-r * T) * si.norm.cdf(d2)
        call_gamma = si.norm.pdf(d1) / (S * sigma * np.sqrt(T))
        call_charm = - si.norm.pdf(d1) * (2*r*T - d2*sigma*np.sqrt(T)) / (2 * T * sigma * np.sqrt(T))
        put = (K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0) - S * si.norm.cdf(-d1, 0.0, 1.0))
        put_delta = -si.norm.cdf(-d1)
        put_theta = - (S * sigma * si.norm.pdf(d1)) / (2 * np.sqrt(T)) + r * K * np.exp(-r * T) * si.norm.cdf(-d2)
        put_gamma = si.norm.pdf(d1) / (S * sigma * np.sqrt(T))
        put_charm = - si.norm.pdf(d1) * (2*r*T - d2*sigma*np.sqrt(T)) / (2 * T * sigma * np.sqrt(T))

        return call, call_delta, call_theta, call_gamma, call_charm
```
